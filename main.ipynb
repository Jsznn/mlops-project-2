#### Feature Engineering

```python
# High-Rate Flag
mean_rate = df['suku_bunga_euribor_3bln'].mean()
df['high_rate_flag'] = (df['suku_bunga_euribor_3bln'] > mean_rate).astype(int)
```

```python
# Sector Grouping
formal = {'manajer', 'pemilik bisnis', 'entrepreneur', 'teknisi', 'sosial media specialis'}
informal = {'pekerja kasar', 'asisten rumah tangga', 'penyedia jasa'}
non_employed = {'mahasiswa', 'pengangguran', 'pensiunan', 'unknown'}

def map_sector(job):
    if job in formal:
        return 'formal'
    elif job in informal:
        return 'informal'
    elif job in non_employed:
        return 'non_employed'
    else:
        return 'other'

df['job_sector'] = df['pekerjaan'].apply(map_sector)
```

```python
cat_columns = df.select_dtypes(include='object').columns.tolist()
print(cat_columns)
```

```python
pekerjaan_mapping = {
    'sosial media specialis': 0,
    'teknisi': 1,
    'pekerja kasar': 2,
    'manajer': 3,
    'asisten rumah tangga': 4,
    'mahasiswa': 5,
    'penyedia jasa': 6,
    'pemilik bisnis': 7,
    'entrepreneur': 8,
    'pengangguran': 9,
    'pensiunan': 10,
    'unknown': 11
}
status_mapping = {
    'menikah': 0,
    'lajang': 1,
    'cerai': 2,
    'unknown': 3
}
pendidikan_mapping = {
    'TIDAK SEKOLAH': 0,
    'Tidak Tamat SD': 1,
    'SD': 2,
    'SMP': 3,
    'SMA': 4,
    'Diploma': 5,
    'Pendidikan Tinggi': 6,
    'unknown': 7
}
gagal_bayar_mapping = {
    'no': 0,
    'yes': 1
}
pinjaman_mapping = { #Pinjaman Rumah & Pinjaman Pribadi
    'no': 0,
    'yes': 1,
    'unknown': 2
}
kontak_mapping = {
    'cellular': 0,
    'telephone': 1
}
bulan_mapping = {
    'mar': 0, 'apr': 1,  # Tidak terdapat data Januari dan Febuari baik pada data train maunpun validasi
    'may': 2, 'jun': 3, 'jul': 4, 'aug': 5, 
    'sep': 6, 'oct': 7, 'nov': 8, 'dec': 9
}
hari_mapping = { # Tidak terdapat data hari sabtu dan minggu
    'mon': 1,
    'tue': 2,
    'wed': 3,
    'thu': 4,
    'fri': 5
}
hasil_mapping = {
    'nonexistent': 0,  # tidak pernah dihubungi
    'failure': 1,
    'success': 2
}
pulau_mapping = {
    'Jawa': 0,
    'Sumatera': 1,
    'Kalimantan': 2,
    'Sulawesi': 3,
    'Bali': 4,
    'NTB': 5,
    'NTT': 6,
    'Papua': 7
}

# Mapping Feature Tambahan
job_sector_mapping = {
    'formal': 0,
    'informal': 1,
    'non_employed': 2
}
```

```python
df['pekerjaan'] = df['pekerjaan'].map(pekerjaan_mapping)
df['status_perkawinan'] = df['status_perkawinan'].map(status_mapping)
df['pendidikan'] = df['pendidikan'].map(pendidikan_mapping)
df['gagal_bayar_sebelumnya'] = df['gagal_bayar_sebelumnya'].map(gagal_bayar_mapping)
df['pinjaman_rumah'] = df['pinjaman_rumah'].map(pinjaman_mapping)
df['pinjaman_pribadi'] = df['pinjaman_pribadi'].map(pinjaman_mapping)
df['jenis_kontak'] = df['jenis_kontak'].map(kontak_mapping)
df['bulan_kontak_terakhir'] = df['bulan_kontak_terakhir'].map(bulan_mapping)
df['hari_kontak_terakhir'] = df['hari_kontak_terakhir'].map(hari_mapping)
df['hasil_kampanye_sebelumnya'] = df['hasil_kampanye_sebelumnya'].map(hasil_mapping)
df['pulau'] = df['pulau'].map(pulau_mapping)


# Mapping Feature Tambahan
df['job_sector'] = df['job_sector'].map(job_sector_mapping)
```

```python
X = df.drop(columns=['berlangganan_deposito', 'customer_number'])
y = df['berlangganan_deposito']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,         
    random_state=23,
    stratify=y
)

print("Jumlah data train:", X_train.shape)
print("Jumlah data test :", X_test.shape)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Train set
sns.countplot(x=y_train, ax=axes[0])
axes[0].set_title("Distribusi Label Train")
axes[0].set_xlabel("Label (Y)")
axes[0].set_ylabel("Jumlah")

# Test set
sns.countplot(x=y_test, ax=axes[1])
axes[1].set_title("Distribusi Label Test")
axes[1].set_xlabel("Label (Y)")
axes[1].set_ylabel("Jumlah")

plt.tight_layout()
plt.show()

print("Distribusi label di train:")
print(y_train.value_counts(normalize=True))

print("\nDistribusi label di test:")
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Logistic Regression
logreg = LogisticRegression(
    penalty='l2',
    C=1.0,
    class_weight='balanced',
    max_iter=200,
    random_state=23
)
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
```

```python
# Evaluasi LogReg
print("=== Logistic Regression ===")
print("Akurasi:", accuracy_score(y_test, y_pred_logreg))
print("f1_score:", f1_score(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))
```

```python
cm = confusion_matrix(y_test, y_pred_logreg)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Tidak Beli', 'Beli'], yticklabels=['Tidak Beli', 'Beli'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Logreg')
plt.show()
```

```python
class_weights = {0: 1, 1:4}
```

```python
# Decision Tree
dtree = DecisionTreeClassifier(
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    class_weight=class_weights,
    random_state=23
)
dtree.fit(X_train, y_train)
y_pred_dtree = dtree.predict(X_test)
```

```python
print("\n=== Decision Tree ===")
print("Akurasi:", accuracy_score(y_test, y_pred_dtree))
print("f1_score:", f1_score(y_test, y_pred_dtree))
print(classification_report(y_test, y_pred_dtree))
```

```python
# Random Forest
rf = RandomForestClassifier(
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    class_weight=class_weights,
    random_state=23
)

rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
print("\n=== Random Forest ===")
print("Akurasi:", accuracy_score(y_test, y_pred_rf))
print("f1_score:", f1_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Tidak Beli', 'Beli'], yticklabels=['Tidak Beli', 'Beli'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
importances = rf.feature_importances_

feature_imp = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False).head(20)

plt.figure(figsize=(10, 6))
sns.barplot(data=feature_imp, x='Importance', y='Feature')
plt.title('Top 20 Feature Importances (Random Forest)')
plt.tight_layout()
plt.show()
```
